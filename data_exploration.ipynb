```python
import pandas as pd
import seaborn as sn
import numpy as np
import matplotlib
```

```python
real_day_generation = pd.read_excel('./data/0590a032-8c8e-4fc7-8c36-2982be2ea28e.xlsx')
real_day_generation
```

```python
energy_program_df = pd.read_excel('./data/PRG260304.xlsx')
energy_program_df
```

```python
thermal_row = energy_program_df[energy_program_df['Unnamed: 2'] == "Térmicas"].index[0]
data_row = energy_program_df.iloc[thermal_row + 1, :] # The "Total" is usually one row below the category header

print(data_row)
```

```python
# Need to get 'Unnamed: 4' to 'Unnamed: 27'
programmed_day_generation = data_row.iloc[4:28]
programmed_day_generation
```

```python
total_excess = np.sum(real_day_generation['Generación Real (MWh)'][0:24]) - np.sum(programmed_day_generation[0:24])
total_excess
```

```python
## we define the morning to be from 6 am to 11 am
morning_excess = np.sum(real_day_generation['Generación Real (MWh)'][6:11]) - np.sum(programmed_day_generation[6:11])
morning_excess
```

```python
#we define the afternoon excess to be from 4 pm to 9 pm
afternoon_excess = np.sum(real_day_generation['Generación Real (MWh)'][16:21]) - np.sum(programmed_day_generation[16:21])
afternoon_excess
```
